## Multi-window feature extraction and statistical analysis

This notebook:

1. Loads physiological signals (EDA, HR, TEMP, ACC) for all subjects
2. Creates stress/rest labels using tags
3. Applies sliding window segmentation (10s, 30s, 60s)
4. Extracts statistical features (mean and standard deviation)
5. Saves the final feature dataset
6. Calculates individual averages per person for rest and stress
7. Creates boxplots of these averages
8. Performs t-tests for each person and for the whole population

```python
# Import required libraries

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import ttest_ind
```

```python
# SETTINGS 

# Main dataset folder
BASE_PATH = r"C:\Users\KARAN\Downloads\Internship Project\Wearable_Dataset\Wearable_Dataset"

# We use the STRESS folder because the tags file gives stress intervals
STRESS_PATH = os.path.join(BASE_PATH, "STRESS")

# Window sizes to compare
WINDOW_SIZES_SEC = [10, 30, 60]

# Sliding window step
# True  = 50% overlap
# False = no overlap
USE_OVERLAP = True
OVERLAP_FRACTION = 0.5

print("Dataset folder:", STRESS_PATH)
```

```python
# Get all subject folders

subjects = sorted([
    d for d in os.listdir(STRESS_PATH)
    if os.path.isdir(os.path.join(STRESS_PATH, d))
])

print("Total subjects found:", len(subjects))
print("Subjects:", subjects)
```

```python
# Small helper to check if a file exists

def must_exist(path):
    return os.path.exists(path) and os.path.isfile(path)
```

```python
# Functions to read Empatica signals


def read_empatica_signal(file_path):
    """
    Read EDA / HR / TEMP signal file.
    Format:
    row 0 = start time
    row 1 = sampling rate
    row 2+ = values
    """
    raw = pd.read_csv(file_path, header=None)

    start_time = raw.iloc[0, 0]
    fs = float(raw.iloc[1, 0])
    values = raw.iloc[2:, 0].astype(float).values

    # Convert start time to datetime
    try:
        t0 = pd.to_datetime(float(start_time), unit="s")
    except:
        t0 = pd.to_datetime(start_time)

    time_index = pd.date_range(
        start=t0,
        periods=len(values),
        freq=pd.Timedelta(seconds=1/fs)
    )

    df = pd.DataFrame({
        "time": time_index,
        "value": values
    })

    return df, fs


def read_acc_magnitude(file_path):
    """
    Read ACC signal and convert X, Y, Z into magnitude.
    """
    raw = pd.read_csv(file_path, header=None)

    start_time = raw.iloc[0, 0]
    fs = float(raw.iloc[1, 0])
    xyz = raw.iloc[2:, :3].astype(float).values

    try:
        t0 = pd.to_datetime(float(start_time), unit="s")
    except:
        t0 = pd.to_datetime(start_time)

    time_index = pd.date_range(
        start=t0,
        periods=len(xyz),
        freq=pd.Timedelta(seconds=1/fs)
    )

    magnitude = np.sqrt(xyz[:, 0]**2 + xyz[:, 1]**2 + xyz[:, 2]**2)

    df = pd.DataFrame({
        "time": time_index,
        "value": magnitude
    })

    return df, fs
```

```python
# Functions for tags and labels


def load_tag_times(tags_path):
    """
    Load tags safely.
    If tags.csv is empty, return an empty datetime series.
    """
    try:
        tags_raw = pd.read_csv(tags_path, header=None)

        if tags_raw.shape[0] == 0:
            return pd.Series([], dtype="datetime64[ns]")

        try:
            tag_times = pd.to_datetime(tags_raw[0].astype(float), unit="s")
        except:
            tag_times = pd.to_datetime(tags_raw[0])

        return tag_times

    except pd.errors.EmptyDataError:
        return pd.Series([], dtype="datetime64[ns]")


def add_labels_to_eda(eda_df, tag_times):
    """
    Default label = rest
    Each pair of tags = one stress interval [start, end]
    If tags are empty, everything stays as rest
    """
    eda_df = eda_df.copy()
    eda_df["label"] = "rest"

    if tag_times is None or len(tag_times) == 0:
        return eda_df

    for i in range(0, len(tag_times), 2):
        if i + 1 >= len(tag_times):
            break

        start = tag_times.iloc[i]
        end = tag_times.iloc[i + 1]

        eda_df.loc[
            (eda_df["time"] >= start) & (eda_df["time"] <= end),
            "label"
        ] = "stress"

    return eda_df
```

```python
# Sliding window feature extraction


def sliding_window_features(subject, eda_df, hr_df, temp_df, acc_df, eda_fs, window_sec, step_sec):
    """
    Create sliding windows and extract simple features:
    - mean
    - standard deviation

    One window = one sample
    """
    rows = []

    window_samples = int(window_sec * eda_fs)
    step_samples = int(step_sec * eda_fs)

    # Safety check
    if window_samples <= 0 or step_samples <= 0:
        return rows

    if len(eda_df) < window_samples:
        return rows

    # Sliding window loop
    for start_idx in range(0, len(eda_df) - window_samples, step_samples):

        end_idx = start_idx + window_samples
        w_eda = eda_df.iloc[start_idx:end_idx]

        start_time = w_eda["time"].iloc[0]
        end_time = w_eda["time"].iloc[-1]

        # Align other signals using time
        w_hr = hr_df[(hr_df["time"] >= start_time) & (hr_df["time"] <= end_time)]
        w_temp = temp_df[(temp_df["time"] >= start_time) & (temp_df["time"] <= end_time)]
        w_acc = acc_df[(acc_df["time"] >= start_time) & (acc_df["time"] <= end_time)]

        # Skip if any signal is missing in the window
        if len(w_hr) == 0 or len(w_temp) == 0 or len(w_acc) == 0:
            continue

        # Window label = majority label in EDA window
        label = w_eda["label"].mode()[0]

        rows.append({
            "subject": subject,
            "window_size_sec": window_sec,

            "eda_mean": w_eda["value"].mean(),
            "eda_std": w_eda["value"].std(),

            "hr_mean": w_hr["value"].mean(),
            "hr_std": w_hr["value"].std(),

            "temp_mean": w_temp["value"].mean(),
            "temp_std": w_temp["value"].std(),

            "acc_mean": w_acc["value"].mean(),
            "acc_std": w_acc["value"].std(),

            "label": label
        })

    return rows
```

```python
# Build the final feature dataset

all_rows = []
skipped_subjects = []

for subject in subjects:
    subject_path = os.path.join(STRESS_PATH, subject)
    print("\nProcessing subject:", subject)

    # Required files
    eda_path = os.path.join(subject_path, "EDA.csv")
    hr_path = os.path.join(subject_path, "HR.csv")
    temp_path = os.path.join(subject_path, "TEMP.csv")
    acc_path = os.path.join(subject_path, "ACC.csv")
    tags_path = os.path.join(subject_path, "tags.csv")

    # Skip subject if any important file is missing
    if not (must_exist(eda_path) and must_exist(hr_path) and must_exist(temp_path) and must_exist(acc_path) and must_exist(tags_path)):
        print("Skipping subject due to missing file:", subject)
        skipped_subjects.append(subject)
        continue

    # Load signals
    eda_df, eda_fs = read_empatica_signal(eda_path)
    hr_df, _ = read_empatica_signal(hr_path)
    temp_df, _ = read_empatica_signal(temp_path)
    acc_df, _ = read_acc_magnitude(acc_path)

    # Load tags and apply labels
    tag_times = load_tag_times(tags_path)
    eda_df = add_labels_to_eda(eda_df, tag_times)

    # Extract features for each window size
    for w in WINDOW_SIZES_SEC:

        if USE_OVERLAP:
            step_sec = max(1, int(w * OVERLAP_FRACTION))
        else:
            step_sec = w

        rows = sliding_window_features(
            subject=subject,
            eda_df=eda_df,
            hr_df=hr_df,
            temp_df=temp_df,
            acc_df=acc_df,
            eda_fs=eda_fs,
            window_sec=w,
            step_sec=step_sec
        )

        all_rows.extend(rows)

    print("Total windows so far:", len(all_rows))

# Final dataframe
features_df = pd.DataFrame(all_rows)

print("\nDone!")
print("Final dataset shape:", features_df.shape)
print("\nLabel counts:")
print(features_df["label"].value_counts())
print("\nWindow size counts:")
print(features_df["window_size_sec"].value_counts())
print("\nSkipped subjects:", skipped_subjects)

features_df.head()
```

```python
# Save the final features dataset

out_file = "features_multisignal_all_windows.csv"
features_df.to_csv(out_file, index=False)

print("Saved file:", out_file)
```

## Person-Level Averages and Boxplots

In this section, we calculate the average physiological values for each person separately for:

- rest
- stress

Then we create boxplots to compare these averages across people.

```python
# Load the saved features dataset


df = pd.read_csv("features_multisignal_all_windows.csv")

print("Dataset loaded successfully")
print("Total rows:", len(df))
print("Total subjects:", df["subject"].nunique())

df.head()
```

```python
# Calculate individual averages per person
# for rest and stress


person_avg = df.groupby(["subject", "label"])[
    ["eda_mean", "hr_mean", "temp_mean"]
].mean().reset_index()

print("Person-level averages created successfully")
person_avg.head(10)
```

```python
# Quick check of person-level averages


print("Total rows in person_avg:", len(person_avg))

print("\nCounts per condition:")
print(person_avg["label"].value_counts())

person_avg.describe()
```

```python
# Boxplot for EDA person averages

plt.figure(figsize=(6, 4))
sns.boxplot(data=person_avg, x="label", y="eda_mean")

plt.title("EDA Average per Person (Rest vs Stress)")
plt.xlabel("Condition")
plt.ylabel("EDA Mean")

plt.show()
```

```python
# Boxplot for HR person averages


plt.figure(figsize=(6, 4))
sns.boxplot(data=person_avg, x="label", y="hr_mean")

plt.title("Heart Rate Average per Person (Rest vs Stress)")
plt.xlabel("Condition")
plt.ylabel("HR Mean")

plt.show()
```

```python
# Boxplot for Temperature person averages


plt.figure(figsize=(6, 4))
sns.boxplot(data=person_avg, x="label", y="temp_mean")

plt.title("Temperature Average per Person (Rest vs Stress)")
plt.xlabel("Condition")
plt.ylabel("Temperature Mean")

plt.show()
```

## T-tests

Now we perform t-tests to check whether there is a statistically significant difference between rest and stress.

We do this in two ways:

1. For each person individually
2. For the whole population

```python
# T-test for each person individually


signals = ["eda_mean", "hr_mean", "temp_mean"]

per_person_ttest_results = []

for subject in df["subject"].unique():

    df_sub = df[df["subject"] == subject]

    for signal in signals:

        rest = df_sub[df_sub["label"] == "rest"][signal].dropna()
        stress = df_sub[df_sub["label"] == "stress"][signal].dropna()

        # We need at least 2 values in each group
        if len(rest) > 1 and len(stress) > 1:

            t_stat, p_value = ttest_ind(rest, stress)

            per_person_ttest_results.append({
                "subject": subject,
                "signal": signal,
                "rest_count": len(rest),
                "stress_count": len(stress),
                "t_statistic": t_stat,
                "p_value": p_value,
                "significant_(p<0.05)": p_value < 0.05
            })

per_person_ttest_df = pd.DataFrame(per_person_ttest_results)

print("Per-person t-test results created")
per_person_ttest_df.head(10)
```

```python
# View all per-person t-test results

per_person_ttest_df
```

```python
# Population-level t-tests

population_ttest_results = []

for signal in signals:

    rest = df[df["label"] == "rest"][signal].dropna()
    stress = df[df["label"] == "stress"][signal].dropna()

    t_stat, p_value = ttest_ind(rest, stress)

    population_ttest_results.append({
        "signal": signal,
        "rest_count": len(rest),
        "stress_count": len(stress),
        "t_statistic": t_stat,
        "p_value": p_value,
        "significant_(p<0.05)": p_value < 0.05
    })

population_ttest_df = pd.DataFrame(population_ttest_results)

print("Population-level t-test results created")
population_ttest_df
```

## Simple interpretation guide

- If **p-value < 0.05**, there is a statistically significant difference between rest and stress.
- If **p-value >= 0.05**, there is no statistically significant difference.

This can be used to interpret EDA, HR, and Temperature differences:
- for each person
- for the full population

```python
# save statistical results

person_avg.to_csv("person_level_averages.csv", index=False)
per_person_ttest_df.to_csv("per_person_ttest_results.csv", index=False)
population_ttest_df.to_csv("population_ttest_results.csv", index=False)

print("Saved:")
print("- person_level_averages.csv")
print("- per_person_ttest_results.csv")
print("- population_ttest_results.csv")
```
